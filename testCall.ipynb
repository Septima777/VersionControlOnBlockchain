```python
import json
import web3

myContractAddress = "0x7E752ac642edCd2460082EBbA8C6E4AEF75Bd186"; 

with open("Data.json") as f:
    info_json = json.load(f)
abi = info_json["abi"]

w3 = Web3(Web3.WebsocketProvider("ws://localhost:8545"))
myContract = w3.eth.contract(address=myContractAddress, abi=abi)

myContract.functions.createNewQuestion("url","body","title","author","editor",1,1,1,1).transact()
```
